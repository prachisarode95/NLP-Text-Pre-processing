# Objective

To demonstrate the following text pre-processing techniques applied on the paragraphs:

1. Stop Word Removal
2. Removing Punctuation Marks
3. Stemming
4. Lemmatization

---
## Step 1: Install and Import Required Libraries

Need to download specific NLTK data packages:
- `punkt` — for tokenization (splitting text into words/sentences)
- `stopwords` — a pre-built list of common English stop words
- `wordnet` — the lexical database used by the Lemmatizer
- `averaged_perceptron_tagger` — for Part-of-Speech tagging used during lemmatization

```python
# Import the NLTK library
import nltk

# Download all required NLTK datasets
nltk.download('punkt')                    # Tokenizer models
nltk.download('punkt_tab')                # Updated tokenizer tables
nltk.download('stopwords')               # English stop words list
nltk.download('wordnet')                 # Lexical database for lemmatization
nltk.download('averaged_perceptron_tagger')   # POS tagger
nltk.download('averaged_perceptron_tagger_eng')  # English POS tagger
nltk.download('omw-1.4')                 # Open Multilingual WordNet
```

```python
# Import specific modules from NLTK
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

---
## Step 2: Define the Input Text

```python
# Define the input: two paragraphs combined as one string

input_text = """
Artificial intelligence is rapidly transforming the healthcare industry; it is enabling doctors,
researchers, and medical professionals to diagnose diseases more accurately! Advanced algorithms
can now analyze medical images, patient records, and clinical data: all within a matter of seconds.
However, is this technology truly reliable? Many experts argue that while AI shows great promise,
it must be carefully validated, tested, and monitored before widespread adoption. The future of
healthcare may indeed depend on how well we integrate these intelligent systems.

Environmental conservation has become one of the most pressing concerns of our time. Forests,
oceans, and wildlife habitats are being destroyed at an alarming rate; the consequences are severe!
Governments, organizations, and individuals must act responsibly: reducing pollution, planting trees,
and protecting endangered species. Can we reverse the damage that has already been done? Scientists
believe that with sustained effort, collaboration, and innovation, it is still possible to restore
balance to our ecosystems. Every action, no matter how small, truly matters.
"""
```

---
## Step 3: Tokenization

```python
# Tokenize the input text into individual words
tokens = word_tokenize(input_text)

# Convert all tokens to lowercase for uniformity
tokens_lower = [token.lower() for token in tokens]

print("Tokens (after lowercasing):", tokens_lower)
print(f"\nTotal number of tokens: {len(tokens_lower)}")
```

---
## Step 4: Removing Punctuation Marks

```python
# Display all punctuation characters that will be removed
print("Punctuation characters :")
print(string.punctuation)
```

```python
# Remove punctuation marks from the token list
# Keep a token only if it is NOT found in the punctuation set

tokens_no_punct = [token for token in tokens_lower if token not in string.punctuation]

print("Tokens (after Punctuation Removal) :", tokens_no_punct)
print(f"\nTotal tokens before punctuation removal : {len(tokens_lower)}")
print(f"Total tokens after punctuation removal : {len(tokens_no_punct)}")
print(f"Punctuation tokens removed : {len(tokens_lower) - len(tokens_no_punct)}")
```

---
## Step 5: Stop Word Removal

```python
# Load NLTK's English stop words list
stop_words = set(stopwords.words('english'))

print("English Stop Words from NLTK:")
print(sorted(list(stop_words))[:40])   # Displaying first 40 stop words alphabetically
print(f"\nTotal number of stop words in NLTK's English list: {len(stop_words)}")
```

```python
# Remove stop words from the punctuation-free token list
tokens_no_stopwords = [token for token in tokens_no_punct if token not in stop_words]

print("Tokens after Stop Word Removal:", tokens_no_stopwords)
print(f"\nTotal tokens before stop word removal: {len(tokens_no_punct)}")
print(f"Total tokens after stop word removal : {len(tokens_no_stopwords)}")
print(f"Stop words removed : {len(tokens_no_punct) - len(tokens_no_stopwords)}")
```

---
## Step 6: Stemming

```python
# Initialize the Porter Stemmer
porter_stemmer = PorterStemmer()

# Apply stemming to each token in the cleaned token list
stemmed_tokens = [porter_stemmer.stem(token) for token in tokens_no_stopwords]

print("Tokens after Stemming (using Porter Stemmer):", stemmed_tokens)
```

```python
# Display a side-by-side comparison of original token vs stemmed form
print("Word-by-Word Comparison: Original Token vs Stemmed Form")
print(f"{'Original Token':<25} {'Stemmed Form':<25}")

for original, stemmed in zip(tokens_no_stopwords, stemmed_tokens):
    # Highlight only those words where the stem differs from the original
    if original != stemmed:
        print(f"{original:<25} {stemmed:<25}")
    else:
        print(f"{original:<25} {stemmed:<25}")
```

---
## Step 7: Lemmatization

```python
# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Convert NLTK POS tags to WordNet POS format
# NLTK tags start with 'J' (adjective), 'V' (verb), 'N' (noun), 'R' (adverb)
# WordNet uses its own constants: ADJ, VERB, NOUN, ADV

def get_wordnet_pos(nltk_tag):
    """
    Mapping an NLTK POS tag to the corresponding WordNet POS constant.
    This helps the lemmatizer produce more accurate results.
    """
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        # Default to noun if the tag is not recognized
        return wordnet.NOUN
```

```python
# Get POS tags for each token in the cleaned token list
pos_tags = nltk.pos_tag(tokens_no_stopwords)

print("POS Tags assigned to each token:", pos_tags)
```

```python
# Apply lemmatization using the POS-aware helper function
lemmatized_tokens = [
    lemmatizer.lemmatize(token, get_wordnet_pos(tag))
    for token, tag in pos_tags
]

print("Tokens after Lemmatization (WordNet Lemmatizer with POS tags):", lemmatized_tokens)
```

```python
# Display a side-by-side comparison of original token vs lemmatized form
print("Word-by-Word Comparison: Original Token vs Lemmatized Form")
print(f"{'Original Token':<25} {'Lemmatized Form':<25}")

for original, lemma in zip(tokens_no_stopwords, lemmatized_tokens):
    if original != lemma:
        print(f"{original:<25} {lemma:<25}")
    else:
        print(f"{original:<25} {lemma:<25}")
```
